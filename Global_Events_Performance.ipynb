# Data Cleaning

-  Replace Gold, Silver, Bronze, Other with 1, 2, 3, 4
-  Replace Male, Female with 1, 2
-  Replace Summer, Winter with 1, 2
-  Change Age to int, missing values = -1

```python
athlete_data['Medal'].replace('Gold', 1, inplace = True)
athlete_data['Medal'].replace('Silver', 2, inplace = True)
athlete_data['Medal'].replace('Bronze', 3, inplace = True)
athlete_data['Medal'].replace(np.nan, 4, inplace = True)

athlete_data['Sex'].replace('M', 1, inplace = True)
athlete_data['Sex'].replace('F', 2, inplace = True)

athlete_data['Season'].replace('Summer', 1, inplace = True)
athlete_data['Season'].replace('Winter', 2, inplace = True)

athlete_data['Age'].replace(np.nan, -1, inplace = True)
athlete_data['Age'] = athlete_data['Age'].astype(int)

athlete_data.sort_values('Year', ascending = True, inplace = True)

athlete_data
```

```python
track_data.head()
```

```python
track_data['Medal'].replace('G', 1, inplace = True)
track_data['Medal'].replace('S', 2, inplace = True)
track_data['Medal'].replace('B', 3, inplace = True)
track_data['Medal'].replace(np.nan, 4, inplace = True)

track_data['Gender'].replace('M', 1, inplace = True)
track_data['Gender'].replace('W', 2, inplace = True)

track_data.sort_values('Year', ascending = True, inplace = True)
track_data
```

```python
#Create a vertical bar chart showing countries that have hosted 

sns.set_style("darkgrid")
athlete_data["Year"] = athlete_data["Year"].astype(str)
hs = athlete_data.groupby("City")["Year"].apply(" , ".join).reset_index()
hs1 = athlete_data["City"].value_counts().reset_index()
h_s = hs.merge(hs1,left_on = "City",right_on = "index",how="left")
h_s2 = h_s.sort_values(by="City_y",ascending=False)

fig = plt.figure(1, figsize=(20,10),frameon=False,dpi=100)
fig.add_axes([0,0,1,1])

ax = sns.barplot("City_y","City_x",
                data=h_s, linewidth=.8,
                palette='summer',
                edgecolor = "k"*h_s["City_x"].nunique())
for i, j in enumerate("Years :"+h_s["Year"]):
    ax.text(.1,i,j,fontsize=12)
plt.grid(True)
plt.xticks(np.arange(0,5,1))
plt.xlabel("count")
plt.ylabel("country")
plt.title("Countries that Hosted Summer Games")
```

## Investigation 1: 1936 Summer Olympics 

This event makes for an intense geopolitical environment for several reasons. WWI ended in 1918, and WWII began in 1939. Economic, social, and political after-effects from the first global war were still largely relevant as countries struggled with poplation disparities, economic depressions, and displaced trade. 

Considering that these Olympics were held in Berlin, Germany directly following Hitler's rise to power in 1933, it is well-documented that the Olympic games became entangled with Nazi politics and intense German nationalism. 

This entanglement resulted in waves of protest as many countries boycotted the IOC's decision to hold the Olympics in Berlin. Regardless, the events went on as planned. Here's how this affected performance.
