## Overview
In this Notebook we want to look at the relationship between GDP and Success in the Olympics (measured in Medal Count) 

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import patsy
import statsmodels.api as sm

# Display plots directly in the notebook instead of in a new window
%matplotlib inline
```

## Data
   - GDP: Historical GDP data
   - athlete_data: Olympic Competetor and Event Info
   - track_and_field_results: Athlete Info and Event Results

```python
# Load Data
GDP = pd.read_csv('../data/gdp_csv.csv')
athlete_data = pd.read_csv('../data/athlete_events.csv', index_col = 0)
track_and_field_results = pd.read_csv('../Data/results.csv', 
                            names = ('Gender', 'Event', 'Location', 'Year', 'Medal', 'Name', 'Nationality', 'Result'))
```

## Data Cleaning
 - Redefine 'athlete_data' to only contain Track & Field events
 - Replace Gold, Silver, Bronze, Other with 1, 2, 3, 4
 - Replace Male, Female with 1, 2
 - Replace Summer, Winter with 1, 2
 - Change Age to int, missing values = -1

```python
# We are only using data from Track and Field Events
athlete_data = athlete_data[athlete_data['Sport'] == 'Athletics']
#track_and_field_results.drop(0, axis=0, inplace = True)
```

```python
athlete_data['Medal'].replace('Gold', 1, inplace = True)
athlete_data['Medal'].replace('Silver', 2, inplace = True)
athlete_data['Medal'].replace('Bronze', 3, inplace = True)
athlete_data['Medal'].replace(np.nan, 4, inplace = True)

athlete_data['Sex'].replace('M', 1, inplace = True)
athlete_data['Sex'].replace('F', 2, inplace = True)

athlete_data['Season'].replace('Summer', 1, inplace = True)
athlete_data['Season'].replace('Winter', 2, inplace = True)

athlete_data['Age'].replace(np.nan, -1, inplace = True)
athlete_data['Age'] = athlete_data['Age'].astype(int)

athlete_data['NOC'].replace('GDR', 'GER', inplace = True)
athlete_data['NOC'].replace('URS', 'RUS', inplace = True)

GDP['Country Code'].replace('DEU', 'GER', inplace = True)

athlete_data.sort_values('Year', ascending = True, inplace = True)
```

## Sparse Out Categorical Information

```python
# Create two DataFrames, recording Gold Medal Count and Total Medal Count
athlete_data_G = athlete_data[athlete_data['Medal'] == 1.0]
athlete_data_Medal = athlete_data[(athlete_data['Medal'] == 1.0) | (athlete_data['Medal'] == 2.0) | 
                                      (athlete_data['Medal'] == 3.0)]

# Record top ten countries in Gold Medal standing and Total Medal standing
GOLD_Top_10 = athlete_data_G['NOC'].value_counts().index[0:10]
Medal_Top_10 = athlete_data_Medal['NOC'].value_counts().index[0:10]

# Track number Gold Count and Total Medal Count for top 10 countries in each category
athlete_data_G_vc = athlete_data_G['NOC'].value_counts()[0:10]
athlete_data_Medal_vc = athlete_data_Medal['NOC'].value_counts()[0:10]

# Limit athlete_data_G and athlete_data_Medal to only contain top 10 countries
athlete_data_G = athlete_data_G[athlete_data_G['NOC'].isin(GOLD_Top_10)]
athlete_data_Medal = athlete_data_Medal[athlete_data_Medal['NOC'].isin(Medal_Top_10)]

# Create two DataFrames, recording GDP of top 10 countries in Gold Medal standings and Total Medal standings
GDP_Top_10_G = GDP[GDP['Country Code'].isin(GOLD_Top_10)]
GDP_Top_10_Medal = GDP[GDP['Country Code'].isin(Medal_Top_10)]

# Create two DataFrames, recording number of participating athletes in top 10 countries in Gold Medal standings 
# and Total Medal standings
athlete_participants_G = athlete_data[athlete_data['NOC'].isin(GOLD_Top_10)]
athlete_participants_Medal = athlete_data[athlete_data['NOC'].isin(Medal_Top_10)]

# Standardize format of GDP in Trillions of USD
def format_value(value):
    return value/1000000000000
GDP_Top_10_G.loc['Value'] = GDP_Top_10_G['Value'].apply(format_value)
GDP_Top_10_Medal.loc['Value'] = GDP_Top_10_Medal['Value'].apply(format_value)

# Print resulting variables
print('Top 10 Gold Medal earning countries:\n\t\t' + str(list(GOLD_Top_10)) + '\n')
print('Top 10 Total Medal earning countries:\n\t\t' + str(list(Medal_Top_10)) + '\n')
print('Athletes in Top 10 Gold Medal earning countries-\n' + str(athlete_participants_G['NOC'].value_counts()) + '\n')
print('Athletes in Top 10 Gold Medal earning countries-\n' + str(athlete_participants_Medal['NOC'].value_counts()) + '\n')
print('Gold Medal counts-\n' + 'NOC:  Count:\n' + str(athlete_data_G_vc) + '\n')
print('Total Medal counts-\n' + 'NOC:  Count:\n' + str(athlete_data_Medal_vc))
```

## Statistical Analysis

### USA - Total Medal count from GDP

```python
# New df to predict from
medals_USA = athlete_data_Medal[athlete_data_Medal['NOC'] == 'USA']['Year'].value_counts(sort=False)
gdp_USA = GDP_Top_10_Medal[GDP_Top_10_Medal['Country Code'] == 'USA']['Value'][-28:]
data_USA = {'medals':medals_USA.values, 'gdp':gdp_USA}
df_USA = pd.DataFrame(data_USA)

medals_USA_G = athlete_data_G[athlete_data_G['NOC'] == 'USA']['Year'].value_counts(sort=False)
gdp_USA_G = GDP_Top_10_G[GDP_Top_10_G['Country Code'] == 'USA']['Value'][-28:]
data_USA_G = {'medals':medals_USA_G.values, 'gdp':gdp_USA_G}
df_USA_G = pd.DataFrame(data_USA_G)


# Ordinary Least Squares linear model, predicting medals from gdp
outcome_USA, predictors_USA = patsy.dmatrices( 'medals ~ gdp', data=df_USA)
mod_USA = sm.OLS(outcome_USA, predictors_USA)
res_USA = mod_USA.fit()

outcome_USA_G, predictors_USA_G = patsy.dmatrices( 'medals ~ gdp', data=df_USA_G)
mod_USA_G = sm.OLS(outcome_USA_G, predictors_USA_G)
res_USA_G = mod_USA_G.fit()


# Print results (in middle section: coef - y intercept of the linear model and slope, P>|t| - p-value stastistic)
#print(res_USA.summary())
#print(res_USA_G.summary())

# Plot the orginal data (as before)
plt.scatter(df_USA['gdp'], df_USA['medals'], label='Data');
plt.scatter(df_USA_G['gdp'], df_USA_G['medals'], label='Data_G');

# Generate and plot the model fit line
html = res_USA.summary().tables[1].as_html()
l = pd.read_html(html, header=0, index_col=0)[0]
x = df_USA['gdp']
# y = slope * x + y-intercept
y = l.coef[1] * x + l.coef[0]
plt.plot(x, y, '-b', linewidth=4, label='Model')

html_G = res_USA_G.summary().tables[1].as_html()
l_G = pd.read_html(html_G, header=0, index_col=0)[0]
x_G = df_USA_G['gdp']
# y = slope * x + y-intercept
y_G = l_G.coef[1] * x_G + l_G.coef[0]
plt.plot(x_G, y_G, '-r', linewidth=4, label='Model_G')

# Plot
plt.xlabel('gdp')
plt.ylabel('medals')
plt.legend()
plt.show()
```

### GBR - Total Medal count from GDP

```python
# New df to predict from
medals_GBR = athlete_data_Medal[athlete_data_Medal['NOC'] == 'GBR']['Year'].value_counts(sort=False)
gdp_GBR = GDP_Top_10_Medal[GDP_Top_10_Medal['Country Code'] == 'GBR']['Value'][-29:]
data_GBR = {'medals':medals_GBR.values, 'gdp':gdp_GBR}
df_GBR = pd.DataFrame(data_GBR)

medals_GBR_G = athlete_data_G[athlete_data_G['NOC'] == 'GBR']['Year'].value_counts(sort=False)
gdp_GBR_G = GDP_Top_10_G[GDP_Top_10_G['Country Code'] == 'GBR']['Value'][-23:]
data_GBR_G = {'medals':medals_GBR_G.values, 'gdp':gdp_GBR_G}
df_GBR_G = pd.DataFrame(data_GBR_G)


# Ordinary Least Squares linear model, predicting medals from gdp
outcome_GBR, predictors_GBR = patsy.dmatrices( 'medals ~ gdp', data=df_GBR)
mod_GBR = sm.OLS(outcome_GBR, predictors_GBR)
res_GBR = mod_GBR.fit()

outcome_GBR_G, predictors_GBR_G = patsy.dmatrices( 'medals ~ gdp', data=df_GBR_G)
mod_GBR_G = sm.OLS(outcome_GBR_G, predictors_GBR_G)
res_GBR_G = mod_GBR_G.fit()


# Print results (in middle section: coef - y intercept of the linear model and slope, P>|t| - p-value stastistic)
#print(res_GBR.summary())
#print(res_GBR_G.summary())

# Plot the orginal data (as before)
plt.scatter(df_GBR['gdp'], df_GBR['medals'], label='Data');
plt.scatter(df_GBR_G['gdp'], df_GBR_G['medals'], label='Data_G');

# Generate and plot the model fit line
html = res_GBR.summary().tables[1].as_html()
l = pd.read_html(html, header=0, index_col=0)[0]
x = df_GBR['gdp']
# y = slope * x + y-intercept
y = l.coef[1] * x + l.coef[0]
plt.plot(x, y, '-b', linewidth=4, label='Model')

html_G = res_GBR_G.summary().tables[1].as_html()
l_G = pd.read_html(html_G, header=0, index_col=0)[0]
x_G = df_GBR_G['gdp']
# y = slope * x + y-intercept
y_G = l_G.coef[1] * x_G + l_G.coef[0]
plt.plot(x_G, y_G, '-r', linewidth=4, label='Model_G')

# Plot
plt.xlabel('gdp')
plt.ylabel('medals')
plt.legend(loc=1)
plt.show()
```
