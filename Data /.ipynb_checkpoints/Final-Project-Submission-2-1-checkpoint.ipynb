# Research Question and Background 
## Research Question: 
What national and personal factors impact performance in the Olympic Games? Standards at the Olympic Games are exceedingly high as Olympic athletes are the best in their sports. The Olympics serves as a meeting point where world class athletes gather to represent their countries in competion. As time progresses, what external factors such as biometrics, GDP, economic recessions, political climates, and drug controversies affect athletic performance at the Summer Olympic games?

## Background
We can distill an expectation of human athletic achievement by comparing the Gold Medal winning attempts at objective sports. For this we have limited our analysis to Track and Field Events in the Olympics because these events have standardized metrics of time and distance(as opposed to subjective scorings such as Gymnastics). From here, we analyze 7 sports for both genders: 100m dash, Mile, Mile Relay, 4x100m Relay, High Jump, Javelin and Shot Put. In these events, we expect a somewhat linear progression in Gold Medal winning attempts accounting for the natural human improvement over time as we train harder and condition our top athletes for peak performance. From this expectantly linear correlation, can we further distill deviations in this progression due to factors that may impact the international climate causing athletes to miss their mark?

# Data Cleaning and Preprocessing
Our Data Cleaning Process consisted of conversion of categorical variables(medals, gender, ) into numerical labels to speed up our dataset queries, and changing values to pandas numeric values.

Events of interest for our project, within the Athletics department: 

- 1500m 
- high jump 
- javelon

```python
#imports 
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import datetime as dt
import patsy
import statsmodels.api as sm

# Display plots directly in the notebook instead of in a new window
%matplotlib inline

#data 
athlete_data = pd.read_csv('athlete_events.csv', index_col = 0)
regional_data = pd.read_csv('noc_regions.csv')
track_data = pd.read_csv('results.csv',  names = ('Gender', 'Event', 'Location', 'Year', 'Medal', 'Name', 'Nationality', 'Result'))
GDP = pd.read_csv('gdp_csv.csv')

#limit athlete_data to Track and Field 
athlete_data = athlete_data[athlete_data['Sport'] == 'Athletics']


#DATA CLEANING: Replace Gold, Silver, Bronze, Other with 1, 2, 3, 4
athlete_data['Medal'].replace('Gold', 1, inplace = True)
athlete_data['Medal'].replace('Silver', 2, inplace = True)
athlete_data['Medal'].replace('Bronze', 3, inplace = True)
athlete_data['Medal'].replace(np.nan, 4, inplace = True)

track_data['Medal'].replace('G', 1, inplace = True)
track_data['Medal'].replace('S', 2, inplace = True)
track_data['Medal'].replace('B', 3, inplace = True)
track_data['Medal'].replace(np.nan, 4, inplace = True)
track_data.drop(0, axis=0, inplace = True)

#DATA CLEANING: Replace Male, Female with 1, 2
athlete_data['Sex'].replace('M', 1, inplace = True)
athlete_data['Sex'].replace('F', 2, inplace = True)

track_data['Gender'].replace('M', 1, inplace = True)
track_data['Gender'].replace('W', 2, inplace = True)

#DATA CLEANING: Replace Summer, Winter with 1, 2
athlete_data['Season'].replace('Summer', 1, inplace = True)
athlete_data['Season'].replace('Winter', 2, inplace = True)

#DATA CLEANING: Change Age to int, missing values = -1
athlete_data['Age'].replace(np.nan, -1, inplace = True)
athlete_data['Age'] = athlete_data['Age'].astype(int)

#DATA CLEANING: Combine all German and Russian Olympic Teams
athlete_data['NOC'].replace('GDR', 'GER', inplace = True)
athlete_data['NOC'].replace('URS', 'RUS', inplace = True)

#DATA CLEANING: Match German Country Code with athlete_data
GDP['Country Code'].replace('DEU', 'GER', inplace = True)

#DATA CLEANING: Display 'Year' data in aescending order 
athlete_data.sort_values('Year', ascending = True, inplace = True)
track_data.sort_values('Year', ascending = True, inplace = True)

#DATA CLEANING: Mapping Track and Field Results File
track_data['Year'] = pd.to_numeric(track_data['Year'])

#DATA CLEANING: Medal Wins and Non-Medal Wins
medal_wins = athlete_data[athlete_data['Medal'] != 4]
non_medal_wins = athlete_data[athlete_data['Medal'] == 4]

#Create seperate dataframes for our events of interest. 
mens_100m = track_data[track_data['Event'] == '100M Men']
womens_100m = track_data[track_data['Event'] == '100M Women']
mens_1500m = track_data[track_data['Event'] == '1500M Men']
womens_1500m = track_data[track_data['Event'] == '1500M Women']
mens_relay = track_data[track_data['Event'] == '4X100M Relay Men']
mens_relay['Result'] = mens_relay['Result'].str.replace('0:', '')
womens_relay = track_data[track_data['Event'] == '4X100M Relay Women']
mens_high_jump = track_data[track_data['Event'] == 'High Jump Men']
womens_high_jump = track_data[track_data['Event'] == 'High Jump Women']
womens_high_jump.at[2166, 'Result'] = 1.97
mens_javelin = track_data[track_data['Event'] == 'Javelin Throw Men']
mens_javelin.at[1263, 'Result'] = 90.30
womens_javelin = track_data[track_data['Event'] == 'Javelin Throw Women']
mens_shot_put = track_data[track_data['Event'] == 'Shot Put Men']
womens_shot_put = track_data[track_data['Event'] == 'Shot Put Men']


#DATA CLEANING: Create Dataframe Including Total Medal Count
athlete_data_Medal = athlete_data[(athlete_data['Medal'] == 1.0) | (athlete_data['Medal'] == 2.0) | 
                                      (athlete_data['Medal'] == 3.0)]
#DATA CLEANING: Record Top Ten Countries in Total Medal Standing
Medal_Top_10 = athlete_data_Medal['NOC'].value_counts().index[0:10]
athlete_data_Medal = athlete_data_Medal[athlete_data_Medal['NOC'].isin(Medal_Top_10)]

#DATA CLEANING: Group 'Team','Year','Medal' into one column
medal_wins['Yearly Medals'] = medal_wins.groupby(['Team', 'Year'])['Medal'].transform('count')

```

**Q4:** How has the height, weight, and age changed over time for both the mens and women's gold medalist in the shot put and 100 meter dash ?

### Men's Shot Put:

```python
M_shotput_data_biometrics = athlete_data[(athlete_data['Event'] == 'Athletics Men\'s Shot Put') & 
                                       (athlete_data['Medal'] == 1)][['Age', 'Height', 'Weight','Event','Year']]

M_shotput_data_biometrics.dropna(inplace=True)

mens_shot_put_wt_slope, mens_shot_put_wt_intercept = np.polyfit( M_shotput_data_biometrics ['Year'], M_shotput_data_biometrics ['Weight'], 1)
mens_shot_put_ht_slope, mens_shot_put_ht_intercept = np.polyfit( M_shotput_data_biometrics ['Year'], M_shotput_data_biometrics ['Height'], 1)
mens_shot_put_age_slope, mens_shot_put_age_intercept = np.polyfit( M_shotput_data_biometrics ['Year'], M_shotput_data_biometrics ['Age'], 1)

print('Slope: ', mens_shot_put_age_slope, ", Intercept: ", mens_shot_put_age_intercept)

m_x = M_shotput_data_biometrics ['Year']
m_y_wt = mens_shot_put_wt_slope * m_x + mens_shot_put_wt_intercept
m_y_ht = mens_shot_put_ht_slope * m_x + mens_shot_put_ht_intercept
m_y_age = mens_shot_put_age_slope * m_x + mens_shot_put_age_intercept

plt.plot(m_x,m_y_wt,label = 'Weight kg')
plt.plot(m_x,m_y_ht, label = 'Height cm')
plt.plot(m_x,m_y_age, label = 'Age yrs')
plt.title('Men\'s')
plt.legend(loc = 2)
plt.xlabel('Year')
plt.show()
```

From this analysis, we can see that since the beggining of the Men's Olympic Shot Put event, the weight and height of the gold medal winner has been increasing. Interestingly, the age has remained relatively the same, around 25 years.

### Women's Shot Put:

```python
W_shotput_data_biometrics = athlete_data[(athlete_data['Event'] == 'Athletics Women\'s Shot Put') & 
                                       (athlete_data['Medal'] == 1)][['Age', 'Height', 'Weight','Event','Year']]

W_shotput_data_biometrics.dropna(inplace=True)

womens_shot_put_wt_slope, womens_shot_put_wt_intercept = np.polyfit( W_shotput_data_biometrics ['Year'], W_shotput_data_biometrics ['Weight'], 1)
womens_shot_put_ht_slope, womens_shot_put_ht_intercept = np.polyfit( W_shotput_data_biometrics ['Year'], W_shotput_data_biometrics ['Height'], 1)
womens_shot_put_age_slope, womens_shot_put_age_intercept = np.polyfit( W_shotput_data_biometrics ['Year'], W_shotput_data_biometrics ['Age'], 1)

print('Slope: ', womens_shot_put_age_slope, ", Intercept: ", womens_shot_put_age_intercept)

w_x = W_shotput_data_biometrics ['Year']
w_y_wt = womens_shot_put_wt_slope * w_x + womens_shot_put_wt_intercept
w_y_ht = womens_shot_put_ht_slope * w_x + womens_shot_put_ht_intercept
w_y_age = womens_shot_put_age_slope * w_x + womens_shot_put_age_intercept

plt.plot(w_x,w_y_wt,label = 'Weight kg')
plt.plot(w_x,w_y_ht, label = 'Height cm')
plt.plot(w_x,w_y_age, label = 'Age yrs')
plt.title('Women\'s')
plt.legend(loc = 2)
plt.xlabel('Year')
plt.show()
```

## Investigation 2: World War I 
### 1920 Summer Games | *Antwerp, Belgium*

Source: https://www.history.com/topics/world-war-i/world-war-i-history

Global conflicts throughout history have had the reputation of drastically disrupting social, economic, and political stability. World Wars have gone so far as to have the power of halting the Olympic Games. 1916 was an example of this phenomenon. The Olympics were set to be held in Berlin, Germany but were canceled amidst the war. 

WWI ended in 1918, and WWII began in 1939. Economic, social, and political after-effects from the first global war were still largely relevant as countries struggled with poplation disparities, economic depressions, and displaced trade. 

As one of the deadliest conflicts in the history of the human race, WWI resulted in an estimated 37 million deaths. Considering that the majority of the deceased were young males, there was a correlational deficit in Olympic involvement in years following the war. This is particuraly evident in the 1920 games, held in Belgium. 

From the point plot "Male Participation over Time", male involvement in the 1920 games was at a local minimum. 

```python
#Create a new dataframe from athlete data for all male Athletics athletes
athlete_data_M = athlete_data[athlete_data['Sex']==1]

insert = athlete_data_M['Year'].value_counts()
fig = plt.figure(figsize=(20,10))
ax = sns.pointplot(x = insert.index,y=insert.values,
                  color='blue')
ax.set_facecolor("white")
sns.set(font_scale=2)
plt.title('Male Participation over Time', color='black')
plt.xticks(rotation = 60)
plt.grid(True, alpha=.2)
plt.ylabel("count", color = 'black')

plt.show()
```

Further historical research can account for possible confounding variables. Obviously, there is a decline in male participation in 1920. While deaths of athletes who were drafted into their respective armies prior to the Olympic Games is one possibility, there are other ways to explore the data. 

Source: https://en.wikipedia.org/wiki/1920_Summer_Olympics

A significant consequence of WWI was the formal division of power based on political standing. This occurance was self-evident in the formation of the Central Powers and Allies, which was a contractual agreement by countries involved to assist one another through the conflict. WWI ended with a victory on behalf of the Allied Powers. As a result, Germany, Austria, Hungary, Bulgaria, and the Ottoman Empire were banned from participating in the Olympics in 1920. 

Poland did not attend as the country was facing the Polish-Soviet War. 

Let's see how total number of countries participating in 1920 compares. 

```python
plt.figure(figsize=(20,10))
plt.subplot(211)
ax = athlete_data.groupby("Year")["NOC"].nunique().plot(kind = "bar",
                                                linewidth = 1,
                                                color = sns.color_palette("Blues"),
                                                edgecolor=  "w"*athlete_data["Year"].nunique())
plt.xticks(rotation = 60)
ax.set_facecolor("white")
plt.ylabel("Count")
plt.title("# of Countries Participating over Time")
plt.grid(True,alpha=.2)
plt.show()
```

Comparatively, 1920 had less overall countries participating than surrounding years, but not by a gap so large as male athlete participation. 

While correlational investigations yield many possibilties for the effects observed in the data, cross-checking with historical sources leads to interesting insights. Global conflict, such as WWI, impacts many aspects of national identity and functionality. The 1920 Olympics experienced a dip in male participants. With the war ending only two years prior, it is feasible that given the historical context this global conflict contributed to the local minimum in terms of male Athletics athlete participation. 

## Investigation 3: The Great Depression

Source: https://en.wikipedia.org/wiki/Great_Depression 

The Great Depression (1929-1939) began in the United States but expanded to become a global economic disater characterized by plummets in stock prices, ~15% decline in worldwide GDP, 50% decline in global trade, reductions in personal income, tax revenue, and profit, increased prices for goods and services, and global unemployment rates as high as 25%-33% .

## Investigation 4: Political Climate

Source: https://en.wikipedia.org/wiki/List_of_Olympic_Games_scandals_and_controversies

Studying the results of the Olympic games in Berlin, Germany provides correlational insight on how political climate affects athletic performance. This event made for an intense geopolitical environment for several reasons. 

The 1936 games directly followed Hitler's rise to power in 1933, and it is well-documented that there was engtanglement with Nazi politics and intense German nationalism. 

It is also certainly feasible that the global economy was experiencing trailing effects from the Great Depression, which began in 1929 and was a relevant obsticale through the latter half of the 1930s.

## Investigation 5: Run Regression on Gold Medal Winning Attempts

```python
sns.set(font_scale=1)

def generate_timed_gold_medal_count_analysis(sport_gold, time_format):  
    year = np.arange(1896, 2016)
    
    sport_gold.Result = (pd.to_datetime(sport_gold.Result, format = time_format ) - pd.to_datetime('1900-01-01'))
    sport_gold.Result = sport_gold.Result.apply(lambda x: pd.Timedelta.total_seconds(x))

    sport_gold_slope, sport_gold_intercept = np.polyfit(sport_gold['Year'], sport_gold['Result'], 1)
    print('Slope: ', sport_gold_slope, ", Intercept: ", sport_gold_intercept)

    pred_result = np.array(sport_gold_slope * year + sport_gold_intercept)
    
    plt.figure()
    plt.xlabel('Year')
    plt.ylabel('Result')
    plt.scatter(sport_gold['Year'], sport_gold['Result'])
    plt.plot(year, pred_result, color = 'red')
    f1 = plt.gcf()
    
def generate_distance_gold_medal_count_analysis(sport_gold):  
    year = np.arange(1896, 2016)
    
    sport_gold.Result = pd.to_numeric(sport_gold.Result)
    
    sport_gold_slope, sport_gold_intercept = np.polyfit(sport_gold['Year'], sport_gold['Result'], 1)
    print('Slope: ', sport_gold_slope, ", Intercept: ", sport_gold_intercept)

    pred_result = np.array(sport_gold_slope * year + sport_gold_intercept)
    
    plt.figure()
    plt.xlabel('Year')
    plt.ylabel('Result')
    plt.scatter(sport_gold['Year'], sport_gold['Result'])
    plt.plot(year, pred_result, color = 'red')
    f1 = plt.gcf()
```

## Investigation 6: Doping 
### 1980 Summer Games | *Moscow, Soviet Union*

In 1976, the testing for anabolic steroids debuted and the anti-doping movement started to gain momentum within the competitive sports realm. However, since the testing was is in its infancy, the testing for the 1980 Olympics was unsuccessful due to the large variations of anabolic steroids that the Soviet Union had synthesized and administered to their athletes. The 1980 Olympics was held in Moscow, Soviet Union (present-day Russia) and at this games, Olympic records were set 241 times and World records were beaten 97 times.

```python
def generate_national_medal_count_analysis(nation):
    medals = medal_wins[medal_wins['NOC'] == nation][['NOC', 'Year', 'Yearly Medals']]
    year = np.arange(1896, 2016)

    medals_slope, medals_intercept = np.polyfit( medals['Year'], medals['Yearly Medals'], 1)
    print('Slope: ', medals_slope, ", Intercept: ", medals_intercept)

    pred_result = np.array(medals_slope * year + medals_intercept)

    plt.figure()
    plt.plot(medals['Year'], medals['Yearly Medals'] , color = 'red')
    plt.title(nation + ' Medal Count Analysis')
    plt.plot(year, pred_result, color = 'blue')
```

```python
# Standardize format of GDP in Trillions of USD
def format_value(value):
    return np.log(value/1000000000000)
GDP['Value'].apply(format_value)

# Plot and compare GDP of Top 10 Medaling Countries
sns.set(font_scale=1)
fig, ax = plt.subplots()
for item in Medal_Top_10:
    ax.semilogy(GDP[GDP['Country Code'] == str(item)]['Year'].values, 
                GDP[GDP['Country Code'] == str(item)]['Value'].values, 
                '-', label=str(item))
    
plt.title('Top 10 Medaling Countries')
fig.legend(loc=1)
plt.xlim(1955, 2025)
plt.xlabel('Year')
plt.ylabel('log(GDP) (in Trillions of USD)')
plt.show()
```

The top ten medaling countries seem to be in a comparable order in respects to their total medal counts. This may suggest a correlation between a countries GDP and its medal potential.

Now we want to create a DataFrame with each Olympic year, each of the top ten medaling countries, their number of athletes and the number of medals won.

```python
# Grabing competitor data and medal counts
Competitors = athlete_data.loc[:, {'NOC', 'Year'}]
Medal_Counts = athlete_data_Medal.loc[:, {'NOC', 'Year'}]

# Creating and empty DataFrame to run a linear model on
predict = pd.DataFrame(columns={'Year', 'Country Code', 'GDP', 'Competitors', 'Medals'})
predict = predict[['Year', 'Country Code', 'GDP', 'Competitors', 'Medals']]

# Populating DataFrame with data
index = 0
for i in range(1960, 2020, 4):
    for el in (Medal_Top_10):
        # Check that GDP exists for this country in this year
        if len(GDP[(GDP['Year'] == i) & (GDP['Country Code'] == el)]['Value']) == 1:
            predict.loc[index] = [i, el, float(GDP[(GDP['Year'] == i) & (GDP['Country Code'] == el)]['Value']), 
                                  float(len(Competitors[(Competitors['Year'] == i) & (Competitors['NOC'] == el)])),
                                  float(len(Medal_Counts[(Medal_Counts['Year'] == i) & (Medal_Counts['NOC'] == el)]))]
        index += 1

# Create an OLS model for our data
outcome, predictors = patsy.dmatrices( 'Medals ~ GDP + Competitors', data=predict)
mod = sm.OLS(outcome, predictors)
res = mod.fit()
print(res.summary())

# Plot the prediction of medals from GDP
sns.regplot(predict['GDP'], predict['Medals'])
plt.title('Medals ~ GDP')
plt.xlabel('GDP')
plt.ylabel('Medals')
plt.show()

# Plot the prediction of medals from number of competitors
sns.regplot(predict['Competitors'], predict['Medals'])
plt.title('Medals ~ Competitors')
plt.xlabel('Competitors')
plt.ylabel('Medals')
plt.show()
```

Both the GDP of a country and the number of athletes sent to the Olympics are significant predictors of the expected medal count.
GDP has a p-value of 0.038 which is <0.05 and Number of Competitors has a p-value of 0.00 which is also <0.05.
Since the p-value for Number of Competitors is lower it is a better predictor of total medals than GDP, but they are both significantly correlated.
